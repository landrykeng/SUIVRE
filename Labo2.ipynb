```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data_facture=pd.read_excel("Data.xlsx", sheet_name="QUESTIONNAIRE_FOSA")
data_cheque=pd.read_excel("Data.xlsx", sheet_name="cheque_List")

data=data_facture.merge(data_cheque, how="left", left_on="_index",right_on="_parent_index")
```

```python
data_facture.columns
```

```python
var=['start', 'end','_start-geopoint_latitude','_start-geopoint_longitude','NOM DE L’ETABLISSEMENT/ACTEUR',
     'NUMERO SÉQUENTIEL DE L’ETABLISSEMENT/ACTEUR', 'Region', 'Departement','Arrondissement','ZONE D’IMPLANTATION DE L’ETABLISSEMENT/ACTEUR',
     'NOM ET CODE EVALUATEUR', 'NOM ET CODE SUPERVISEUR ',
       "DATE DE L'ENQUETE ",'Fonction du répondant','Catégorie de la formation sanitaire','Type de Gestionnaire de l’établissement/acteur',
       'FOSA', 'Annee', 'Mois', 'Numéro de facture','Statut de Facture selon le Médecin Conseil',
       'Categorie', 'Nombre de chèque','Numéro du Cheque','Nom Bénéficiaire', 'Valeur total mensuelle (en FCFA)',
       'Chèque Validé', 'MOTIFS DE REJETS','Patient référencé'
     ]

good_data=data[var]
```

```python
good_data=good_data.rename(columns={
 "_start-geopoint_latitude": "Latitude",
 "_start-geopoint_longitude":"Longitude",
 "NOM DE L’ETABLISSEMENT/ACTEUR": "Etablissement",
 "NUMERO SÉQUENTIEL DE L’ETABLISSEMENT/ACTEUR": "Num_seq",
 "ZONE D’IMPLANTATION DE L’ETABLISSEMENT/ACTEUR":"Zone",
 "NOM ET CODE EVALUATEUR":"Evaluateur",
 'NOM ET CODE SUPERVISEUR ': "Superviseur",
 "DATE DE L'ENQUETE ":"Date" ,
 'Fonction du répondant':"Fontion_repondant",
 'Catégorie de la formation sanitaire':"Categorie_formation",
 'Type de Gestionnaire de l’établissement/acteur':"Type_gestionnaire",
 'Numéro de facture':"Num_facture",
 'Statut de Facture selon le Médecin Conseil':"Statut_facture",
 'Nombre de chèque':"Nb_cheque",
 'Numéro du Cheque':"Num_cheque",
 'Nom Bénéficiaire':"Bénéficiaire",
 'Valeur total mensuelle (en FCFA)':"Montant_mensuel",
 'Chèque Validé':"Statut_cheque",
 'MOTIFS DE REJETS': "Motif_rejet",
 'Patient référencé':"Patient_ref" 
})
good_data["Date"]=good_data["Date"].dt.date
good_data["Statut_cheque"]=good_data["Statut_cheque"].replace({"Oui":"Validé", "Non":"Non Validé"})
good_data["Coherence"]=np.where(((good_data["Statut_facture"]=="Acceptée") & (good_data["Statut_cheque"]=="Validé")) | ( (good_data["Statut_facture"]=="Rejetée") & (good_data["Statut_cheque"]=="Non Validé")),"Coherent","Non Coherent")
good_data
```
